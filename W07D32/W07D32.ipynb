```python
##### code by student 3
########################

class bike():
    
    def __init__(self,color,name):
        self.color=color
        self.name= name
        self.maxspeed=100
        self.speed=0
        self.petrol=0
        self.maxpetrol=10
    def increase_speed(self):
        if self.petrol==0:
            print("Fill the petrol first")
        else:
            if self.speed+10>self.maxspeed:
                print("For safty reason bike speed can not increased than 100km/hr")
            else:
                self.speed+=10
                print("Speed is increased by 10km/hr", self.speed)
    def decrease_speed(self):
        if self.petrol==0:
            print("Petrol tank is empty you eventally gonna stop!")
        else:
            if self.speed == 0:
                print("Speed can not be in negative")
            else:
                self.speed-=10
                print("Speed is decreased by 10km/hr")
    def fill_petrol(self):
        if self.petrol+1>self.maxpetrol:
            print("Petrol is full")
        else:
            self.petrol+=1
            print("your tank has now this much fuel:",self.petrol)
```

### Single Inheritance

```python


class Animal:
    
    def __init__(self,name,age):
        self.name = name
        self.age = age
        
    def eat(self):
        print(self.name,"is eating")
        
A1 = Animal('mithu',2)
A1.eat()


class Dog(Animal):
    
    breed = None
    
    def bark(self):
        print(self.name,"is barking")
        
    def eat(self):
        print(self.name,"is EATING")
        
D1 = Dog('bruno',3)

D1.bark()

D1.eat()
        
        
```

### multilevel inheritance

```python
# Child and grandfather


class Labrador(Dog):
    
    Dog.breed = 'Labra'
    
    def walk(self):
        
        print(self.name,'is walking and is smart')
        
    def eat(self):
        
        print('please slow down',self.name)
        
L1 = Labrador('luke',5)

L1.eat()
L1.walk()
L1.bark()
```

### mutiple inheritance

```python
class p2:
    
    def __init__(self):
        print('in P2')

class p1:
    
    def __init__(self):
        print('in P1')
        
    def eat(self):
        
        print('eat with p1')
           
class p3:
    
    def __init__(self):
        print('in P3')
        
    def eat(self):
        
        print('eat with p3')
        
        
class child(p1,p3,p2):
    
    def __init__(self):
        
        print('in child')
```

```python
#help(child)

C1 = child()
C1.eat()
```
