#### timedelta

```python
this calss is used for calculations of difference or any manipulation of dates
```

```python
from datetime import timedelta

TD = datetime.now()
T1D = TD - timedelta(10)

print(TD,'\n',T1D)
```

```python
#parameters for timedelta
#days
#hours
#minutes
#seconds
#weeks

TD = datetime.now()
T1H = TD - timedelta(hours=1)

print(TD,'\n',T1H)
```

```python
TD = datetime.now()
T1W = TD - timedelta(weeks=1)

print(TD,'\n',T1W)
```
