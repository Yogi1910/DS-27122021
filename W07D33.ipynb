```python
### Code by student 1

class Vehicle:
    
    def __init__(self,brand,model,color):
        self.brand=brand
        self.model=model
        self.color=color
        self.key=False
        self.total_time=0
        self.mileage=18
        self.speed=0
        

class car(Vehicle):
        
    def start(self):
        if self.key==False:
            self.key=True
            print(self.brand,self.model,'is started and drive safely')
        else:
            print(self.brand ,self.model,'is already started. Good to go')
    
    def stop(self):
        if self.key==False:
            print(self.brand ,self.model,'is already off.')
        else:
            self.key=False
            print(self.brand ,self.model,'is off and see you soon')
            
            
    def run_time(self):
        if self.key==False:
            print('please start the', self.brand ,self.model,'car')
        else:
            self.total_time+=10
            print('total time run in minutes : ',self.total_time)

    def inc_speed(self):
        if self.key==True:
            self.speed+=10
            print('current speed is',self.speed)
            if self.speed>80:
                print('you have reached max limit and drive carefully')       
        else:
             print('please start the', self.brand ,self.model,'car') 

    def dec_speed(self):
        if self.key==True:
            self.speed-=10
            if self.speed<0:
                print('car is stopped. No more decrease is possible')
            else:
                print('currrent speed is',self.speed)        
        else:
             print('please start the', self.brand ,self.model,'car')
                
                
    def drive_details(self):
        if self.key==False:
            print('please start the', self.brand ,self.model,'car')
        else:
            distance=(self.speed*self.total_time)/60
            fuel_cost=(100*distance)/self.mileage
            print('total distance travelled in km : ',distance)
            print('total fuel cost in Rs : ',fuel_cost)
            
            
```

```python
class bookmymovie:
    
    
    def __init__(self):
        
        self.row = int(input('how many rows in the house'))
        self.col = int(input('how many columns in the house'))
        self.no_of_seats = self.row*self.col
        self.matrix = []
        self.seat_count = 0
        self.current_income = 0
        self.total_income = 0
        self.u_details = {}
        
        for i in range(self.row):
            a = []

            for j in range(self.col):
                a.append('S')

            self.matrix.append(a)
        print(self.matrix)
        
        
    def show_seats(self):
        
        print('\nCinema')
        print(end='   ')
        
        for j in range(1,self.col+1):
            print(j,end = " ")
          
        print(end= '\n')
            
        a = 0
        
        for i in self.matrix:
            a+=1
            print(a,end = "  ")
            print(" ".join(i))
            
            
    def buy(self):
        
        a = int(input('Enter the row you want'))
        b = int(input('Enter the col you want'))
        
        if self.matrix[a-1][b-1]=='B':
            
            print('The seat is already booked')
            
            #return
        
        elif self.no_of_seats <60:
            
            price = 10
            
            print('ticket per person is $10,do you want to continue')
            
        elif self.no_of_seats >=60:
            
            pass
            
        
        
        else:
            price = 8 
            
            print('ticket per person is $8,do you want to continue')
            
            
        user_input = input()
        
        if user_input == 'y' or user_input == 'Y':
            
            u_dict = {}
            Uname = input('enter your name')
            Ugen  = input('enter your gender')
            UAge = input('enter your age')
            UPN  = input('enter your mobile no')
            
            self.matrix[a-1][b-1] = "B"
            self.seat_count +=1
            self.current_income +=price
            u_dict[(a,b)] = [Uname,Ugen,UAge,UPN,price]
            
            
            self.u_details.update(u_dict)
            
            print('Booked Successfully')
        else:
            
            print('Booking could not be completed')
            
            
    def info(self):
        
        check_a = int(input('Tell us the row you booked'))
        check_b = int(input('Tell us the col you booked'))
        
        if self.matrix[check_a-1][check_b-1]=="B":
            
            user = self.u_details[check_a,check_b]
            
            print('name',user[0])
            print('gender',user[1])
            print('age',user[2])
            print('PNo',user[3])
            print('Price',user[4])
            
            
        else:
            
            print('Sorry we could not find your seat')
            
            
    #def total_revenue()
    #def stats()
    
    def menu(self):
        
        ans = True
        
        while ans:
            
            print('\nEnter the choice corresponding to the number')
            
            self.choice= int(input('\n 1. show the seats \n 2. Buy ticket \n 3. Statistics \n 4. showbooked tickets'))
            
            if self.choice ==1:
                self.show_seats()
                
            elif self.choice ==2:
                self.buy()
                
            elif self.choice ==3:
                pass
                
            elif self.choice ==4:
                self.info()
                
            else:
                ans= False
                print('not a valid choice')
```
