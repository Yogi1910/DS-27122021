```python
import requests,csv

url = 'https://jsonplaceholder.typicode.com/todos'

r = requests.get(url)

r.status_code

r1 = r.json()


```

```python
data = []
for item in r1:
    data.append([item['userId'],item['id'],item['title'],item['completed']])
    
data
    
    

with open('JSON.csv','w') as f:
    CW = csv.writer(f)
    CW.writerow(['userid','id','title','status_completed'])
    for i in data:
        CW.writerow(i)
    
    

```
