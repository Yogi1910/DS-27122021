#### continue

```python
# We want to stope a particular iteration

# Q1: Print all the numbers from 1 to 50 skipping numbers that are divided by 3 and 5.


# pass and continue are working same here
for x in range(1,51):
    
    if x %3==0 and x%5==0:  
        
        
        continue
    else:
        print(x,end=' ')
```
