```python
import re

class book_my_show:
    
    def __init__(self):
        self.row=int(input('enter the number of rows : '))
        self.column=int(input('enter the number of column : '))
        self.total_seats=self.row*self.column
        self.matrix=[]
        self.seat_count=0
        self.current_income=0
        self.total_income=0
        self.user_details={}
        
        for i in range(self.row):
            a=[]
            for j in range(self.column):
                a.append('S')
            self.matrix.append(a)
        print(self.matrix)
        
    def show_seats(self):
        print('\n')
        print('Cinema: ')
        print(end='  ')
        for i in range(1,self.row+1):
            print(i,end=' ')
        print('')
        
        a=1
        for j in self.matrix:
            print(a,end=' ')
            print(' '.join(j))
            a+=1
            
    def buy(self):
        try:
            print('\n')
            row_selection=int(input('enter a row your looking for : '))
            column_selection=int(input('enter a column your looking for : '))

            if self.matrix[row_selection-1][column_selection-1]=='B':
                print('already booked')

            elif self.total_seats<60:
                print('\n')
                price=10
                print('price per person is 10$')
                print('please enter y if you wish to continue')

            elif self.total_seats>=60:
                print('\n')
                price=8
                print('price per person is 10$')
                print('please enter y if you wish to continue')

            else:
                pass

            user_input=input().capitalize()
            if user_input=='Y':
                print('\n')
                print('please enter the following details : ')
                u_details={}
                
                name=input('enter the name : ')
                
                while True:
                    gender=input('enter your gender : ').lower()
                    if gender=='male' or gender=='female':
                        break
                    else:
                        print('\n')
                        print('mention proper gender please')
                        continue
                    
                age=input('enter your age : ')
                
                while True:
                    mobile_number=input('enter the mobile number : ')
                    if re.search('[0-9]{10}',mobile_number):
                        break
                    else:
                        print('enter the proper 10 digit mobile number')
                        continue
                    
                self.matrix[row_selection-1][column_selection-1]='B'
                u_details[(row_selection,column_selection)]=[name,gender,age,mobile_number]
                self.user_details.update(u_details)
                
                self.seat_count+=1
                self.current_income+=price
                print('\n')
                print('ticket is successfully booked')

            else:
                print('enter the proper details')
        except:
            print('\n')
            print('enter the proper row and column')

    def statistics(self):
            print('\n')
            print('the number of purchased tickets : ',self.seat_count)
            print('percentage of tickets sold : ',(self.seat_count*100)/self.total_seats,'%')
            print('current income : {}'.format(self.current_income))
            print('total income : ',self.total_seats*10)

    def user_info(self):
        try:
            row_choice=int(input('enter the row you booked : '))
            column_choice=int(input('enter the column you booked : '))

            if self.matrix[row_choice-1][column_choice-1]=='B':
                user=self.user_details[(row_choice,column_choice)]
                print('\n')
                print('name : ',user[0])
                print('gender : ',user[1])
                print('age : ',user[2])
                print('mobile_number :',user[3])

            else:
                print('\ndetails not found')
        except:
            print('\nenter the proper row and column')

    def menu(self):
        booking=True 
    
        while booking:  
            print('\n')
            print('1. show the seats')
            print('2. buy a ticket')
            print('3. statistics')
            print('4. show the booked tickets user info')
            print('0. Exit')
            print('\n')

            choice=int(input('enter your choice : '))

            if choice==1:
                self.show_seats()
            elif choice==2:
                self.buy()
            elif choice==3:
                self.statistics()
            elif choice==4:
                self.user_info()
            elif choice==0:
                print('see you soon')
                break
            else:
                print('\ninvalid choice')
                
```

```python
obj = book_my_show()

obj.menu()
```

### setattr() getattr()

```python
###  set and get attributes with help of two functions

# getattr(object,name,default)---# so default value will be printed out if name is not available

# setattr(object, name, value)


class Person:
    
    name = 'Yogesh'
    
P = Person()
P.name
 









```

```python
getattr(P,'name')

getattr(P,'name',45)

setattr(P,'name','Sachin')

print(P.name)

setattr(P,'age',55)

print(P.age)

```

### __dict__

```python
class Person:
    
    species = 'HS'
    
    def __init__(self,name):
    
        self.name = name
    
P = Person('Adam')

print(P.__dict__)

setattr(P,'gender','M')

print(P.__dict__)
 
    
    
```
