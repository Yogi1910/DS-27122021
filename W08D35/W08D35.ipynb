```python
import csv


def main_menu():
    print("Welcome to Dating App Tinder")
    while True:
        print("""
        ----MAIN MENU----
        FIND YOUR LOVE 
        1. REGISTER
        2. LOGIN 
        3. EXIT""")
        
        x = int(input("Enter your choice: "))
        if (x == 1):
            register()
            break
        elif (x == 2):
            login()
            break
        elif (x == 3):
            print("Good bye see you next time....")
            break
            
def register():
    print("----Register----")
    
    with open("dating_data.txt","a") as f:
        write = csv.writer(f)
        
        first_name = input("Enter your first name: ")
        last_name = input("Enter your last name: ")
        user_name = first_name + last_name[0] +"-tinder"
        print("Your system  generated user name is: ",user_name)
        email = input("Enter your email :")
        password = input("Enter your password:")
        gender = input("Enter your gender:")
        dob = input("Enter your date of birth : dd/mm/yy: ")
        tagline = input("Enter your tagline :")
        about = input("Tell us about you: ")
        intrest = ["Food", "Cricket", "Art", "Reading", "Travelling", "Bikes", "Cars"]
        print(intrest)
        select_intrest = input("Choose your intrest from the list")
        
        write.writerow([first_name,last_name,user_name,email,password,gender,dob,tagline, about,select_intrest])
        print("Data saved to file ")

def login():
    print("----Login----")
    not_log_in = True
    while (not_log_in == True):
        with open("dating_data.txt","r") as f:
            user_name = input("Enter your User name :")
            password = input("Enter your password: ")
            reader = csv.reader(f)
            for row in reader:
                for col in row:
                    if (col == user_name and row[4] == password):
                        not_log_in = False
                    else:
                        break
                    
            if (not_log_in == True):
                print("Wrong username or  password! Try Again")
            else:
                print("Access granted Happy Dating")
                
```

## bs4 BeautifulSoup

```python
from bs4 import BeautifulSoup

import html5lib

import csv

```

```python
URL = 'https://thegreatestbooks.org/'

r = requests.get(URL)

r.content

soup = BeautifulSoup(r.content,'html5lib')  # html.parser

soup





```

## find() and find_all()

```python
soup1 = soup.find_all('li', class_ = 'item pb-3 pt-3 border-bottom')

soup1

books = []

for book in soup1:
    b = book.find('h4')
    b = b.text
    #print(b.text)
    b = b.replace('\n','')
    b= b.strip()
    books.append()
    
    
    
books
```

```python
soup1 = soup.find_all('li', class_ = 'item pb-3 pt-3 border-bottom')

soup1

books = []

for book in soup1:
    b = book.find('h4')
    b = b.text
    #print(b.text)
    b = b.replace('\n','')
    b= b.strip()
    books.append(b.split('.',1))
    
    
    
    
    
books

with open('movies.csv','w') as f:
    CW = csv.writer(f)
    
    CW.writerow(['no','movie'])
    CW.writerows(books)
    
    
```
